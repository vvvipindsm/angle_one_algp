### Imports

```python
import numpy as np
import pandas as pd
from stratmanager import StrategyManager
import sklearn.mixture as mix
from ta.momentum import RSIIndicator
import matplotlib.pyplot as plt
import yfinance
```

```python
stock = "LT.NS"

data = yfinance.download(tickers = stock,start="2022-01-01",end="2022-06-01",
                        interval = "1h", group_by = 'ticker', auto_adjust = True)
data = data[["Open", "High", "Low", "Close", "Volume"]]


data1 = yfinance.download(tickers = stock,start="2022-06-01",end="2023-01-01",
                        interval = "1h", group_by = 'ticker', auto_adjust = True)
data1 = data1[["Open", "High", "Low", "Close", "Volume"]]


data3 = yfinance.download(tickers = stock,start="2023-01-01",
                        interval = "1h", group_by = 'ticker', auto_adjust = True)
data3 = data3[["Open", "High", "Low", "Close", "Volume"]]
```

```python
new_data = pd.concat([data,data1,data3])
```
