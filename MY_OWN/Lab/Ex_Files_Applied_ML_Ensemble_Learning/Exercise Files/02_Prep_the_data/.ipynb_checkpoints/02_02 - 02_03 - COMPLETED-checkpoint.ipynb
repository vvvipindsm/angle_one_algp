## Prepare Data: Clean the data

_In this lesson we're going to quickly clean up the continuous features and then in the next video we will clean up the categorical features. Once again, we are doing a very quick pass over this aspect just to get the data in better shape to fit a model on it, if you wanted to learn more about this aspect of machine learning you should take my Feature Engineering course in this Applied Machine Learning series._

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition (we are only using the training set).

This dataset contains information about 891 people who were on board the ship when departed on April 15th, 1912. As noted in the description on Kaggle's website, some people aboard the ship were more likely to survive the wreck than others. There were not enough lifeboats for everybody so women, children, and the upper-class were prioritized. Using the information about these 891 passengers, the challenge is to build a model to predict which people would survive based on the following fields:

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

### Read in Data

_So we will read in our full dataset just like we did last time but we are going to import a few additional packages that will help us clean up all of our data and visualize that data._

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('../titanic.csv')
titanic.head()
```

### Clean continuous variables

_So the first step is going to be filling in the missing values for the `Age` feature. Now, if we run this `.isnull().sum()` command, that will essentially just count up all of the missing values for each feature in our data. You can see that are 177 missing values for `Age` and if you explore this data, you'll find that it is missing at random._

_You'll also notice that there are missing values for `Cabin` and `Embarked` - we will handle those in the next lesson since they are categorical features._

#### Fill missing for `Age`

```python
titanic.isnull().sum()
```

_Returning to `Age`, since it's missing at random, we want to just replace it with a value that won't sway the model one way or the other (since missing `Age` doesn't mean anything in itself). So we're going to replace the missing values by calling this `.fillna()` method on titanic['Age']. So we will just tell it that we want to fill those nas with the mean age value by calling titanic[`Age`] and calling the mean method. Lastly, we pass in the `inplace` argument to tell it to alter `titanic` as it stands now without require a new dataframe to be called._

_Now if we just print out the top 10 rows in the dataframe, we can see in the age column that these are all integers except for 26.699118 so that is the replacement for a missing value_

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.head(10)
```

#### Combine `SibSp` & `Parch`

_Based on the definition of `SibSp` and `Parch` we know that both have to do with family size (one is the number of siblings and spouses, the other is the number of parents and children)._

_I'm going to quickly print out these categorical plots that will show the average survival rate by different values of these features. What we're doing here is we're looping through these two features, telling matplot lib to create a NEW plot for each, then we're calling the catplot method from seaborn and telling it to create a point plot using the titanic data set and the `Survived` column and then which ever column we have from this loop._

_The vertical line is basically an error bar that corresponds to the sample size at each level (larger bar = smaller sample)._

_The key to notice here is that both features have the same response to survival rate. As you have more siblings and spouses or more parents and children, you were less likely to survive._

_Since these features are so closely related, we can combine them into one feature._

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic, kind='point', aspect=2, )
```

_Combining them really is as simple as just just adding the two columns together and storing it as `Family cnt`. So just assign `Family cnt` to titanic['SibSp'] plus titanic['Parch']._

```python
titanic['Family_cnt'] = titanic['SibSp'] + titanic['Parch']
```

#### Drop unnnecessary variables

_The last thing we will do for the continuous features is we will drop `SibSp` and `Parch`. The value that these two features are providing will be captured by the family count feature. So they will become redundant if and leaving them all in together could create some multi-collinearity issues. Additionally, we will drop the `PassengerId` feature because it's an identifier with no distinguishing power on whether somebody survived or not._

_All we need to do to drop them is just pass them in as a list, indicate to drop them as columns using `axis=1` and alter titanic in place. Then we'll print out the top 5 rows to just look at what our dataset looks like._

_You can see we have a `Family cnt` feature while `SibSp`, `Parch`, and `PassengerId` are all gone. Now we can see that there's still some categorical features that need to be cleaned up and we will address that in the next lesson._

```python
titanic.drop(['PassengerId', 'SibSp', 'Parch'], axis=1, inplace=True)
```

```python
titanic.head()
```

### Clean categorical variables

_In this lesson we will continue the cleaning we started in the last section but now we will focus on categorical features._

#### Fill in missing & create indicator for `Cabin`

_Lets start by creating an indicator for `Cabin`._

_We we doing this? Well, as a quick reminder running this same `isnull().sum()` method to count the missings. We see that `Cabin` is missing for 687 people in this dataset. Recall for `Age` we simply replaced the missing values with the average value for `Age`. We were able to take that approach because `Age` was missing at random so we couldn't really bake any information into that missing value._

```python
titanic.isnull().sum()
```

_It's not the same in this case. Lets take a quick look at survival rate based on whether `Cabin` is missing in this dataset or not. We'll do that by calling a group by on titanic, and then we will tell it to group based on titanic['Cabin'].isnull() - that will just return a True (missing) or False (not missing) for each value of `Cabin`. Then lets tell it to get the mean Survival rate for each group._

_We can see that the survival rate is **drastically** lower when `Cabin` is missing (True here). So that would lead us to believe that `Cabin` was NOT missing at random, it in fact meant that these people DIDN'T have a cabin. As a result, they likely were sleeping somewhere on the ship that made it VERY difficult for them to escape. We want to make sure we capture that information in the way that we encode this variable._

```python
titanic.groupby(titanic['Cabin'].isnull())['Survived'].mean()
```

_So we're going to create a really simple Cabin indicator (had a Cabin or not) using the `where()` method from numpy that acts as an if statement. So the condition we want to use is whether `Cabin` is null, so titanic[`Cabin`].isnull(). If it is, return a 0, if not return a 1. And we're going to store that in `Cabin ind`. And lets print out the top 5 rows to make sure it worked._

_Sure enough wherever `Cabin` is missing, `Cabin ind` is a 0._

```python
titanic['Cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
titanic.head()
```

#### Convert `Sex` to numeric

_Next we need to convert the `Sex` feature from `male` vs `female` to numeric. The reason we do this is because a model doesn't know what `male` vs `female` really means, it just knows there are two values for `Sex`. What those values are doesn't really matter - so if we convert it to being numeric it just makes it a little easier for most models to handle._

_So we have a dictionary that does the gender to numeric mapping and now we just need to apply that to the `Sex` column using the `.map()` method in pandas. To map it we call titanic[`Sex`], then the `.map()` method and we tell is to use the `gender num` dictionary to control the mapping. So run that and print out the first few rows. You can see that now instead of `male` and `female` it is 0 and 1._

```python
gender_num = {'male': 0, 'female': 1}

titanic['Sex'] = titanic['Sex'].map(gender_num)
titanic.head()
```

#### Drop unnecessary variables

_Lastly, we want to drop unnecessary variables to clean up our data. Now that we have `Cabin` indicator, we don't need the raw `Cabin` feature because it becomes redundant. `Name` and `Ticket` are both identifiers but don't really have any determining power on whether somebody survived or not._

_So the only feature we're dropping that we haven't discussed is `Embarked`. If you explore this datasets you will find that there are different survival rates with the different levels of teh `Embarked` feature. However, this was not a causal relationship. What you'll find is that different % of people boarding at each port were women and different % of people boarding at each port had cabins. So `Sex` and `Cabin Indicator` are the real causal features here, NOT the embarked feature. So we're going to go ahead and drop this feature as well._ 

_So we will run this cell and print out the top 5 rows. So now we can see we have a nice clean dataset without any irrelevant or repetitive variables and without any missing values._

```python
titanic.drop(['Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
titanic.head()
```

### Write out cleaned data

_Lastly, lets just write out this cleaned dataset so we can pick it up in the next lesson and split it into our training, validation, and test sets._

```python
titanic.to_csv('../titanic_cleaned.csv', index=False)
```
