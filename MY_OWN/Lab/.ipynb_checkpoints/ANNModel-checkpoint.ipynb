```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
```

```python
# Function to get historical stock data from Yahoo Finance API
def get_stock_data(ticker, start_date):
    data = yf.download(ticker, start=start_date)
    return data
```

```python
# Function to create sequences for time series prediction
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        seq = data[i : i + sequence_length]
        sequences.append(seq)
    return np.array(sequences)
```

```python
# Function to build and train the ANN model
def build_ann_model(input_dim):
    model = Sequential()
    model.add(Dense(64, input_dim=input_dim, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
stock_ticker = "AAPL"  # Apple Inc. stock ticker
start_date = "2010-01-01"
end_date = "2021-12-31"
sequence_length = 10

# Get stock data
data = get_stock_data(stock_ticker, start_date)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close_dime = scaler.fit_transform(data[['Close', 'High']])
scaled_data = MinMaxScaler(feature_range=(0, 1))
scaled_close_data = scaled_data.fit_transform(data[['Close']])
```

```python
sequences = create_sequences(scaled_close_data, sequence_length)
```

```python
# Split data into training and testing sets
train_size = int(len(sequences) * 0.8)
test_size = len(sequences) - train_size
train_data, test_data = sequences[:train_size], sequences[train_size:]
```

```python
def handleData(data):
    print(data.shape[0])
    lendf = data.shape[0] -1 
    if data[lendf - 1] > data[lendf]:
        return 1
    else:
        return 0
    return 0
squared_numbers = list(map(handleData, train_data))
dd = np.array(squared_numbers)
dd
```

```python
# Separate input (X) and output (y) variables
X_train, y_train = train_data[:, :-1], train_data[:, -1]
X_test, y_test = test_data[:, :-1], test_data[:, -1]
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
input_dim = X_train.shape[1] * X_train.shape[2] 
input_dim
```

```python
model = build_ann_model(input_dim)
model.fit(X_train.reshape(-1, input_dim), y_train[:, 0], epochs=300, batch_size=32, verbose=1)
```

```python
# Evaluate the model
train_loss = model.evaluate(X_train.reshape(-1, input_dim), y_train[:, 0], verbose=0)
test_loss = model.evaluate(X_test.reshape(-1, input_dim), y_test[:, 0], verbose=0)
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)
```

```python
# Make predictions
train_predict = model.predict(X_train.reshape(-1, input_dim))
test_predict = model.predict(X_test.reshape(-1, input_dim))
```

```python
# Inverse transform predictions

train_predict = scaled_data.inverse_transform(train_predict)
y_train = scaler.inverse_transform(y_train.reshape(-1, y_train.shape[1]))
test_predict = scaled_data.inverse_transform(test_predict)
y_test = scaler.inverse_transform(y_test.reshape(-1, y_train.shape[1]))
#train_predict.shape
```
