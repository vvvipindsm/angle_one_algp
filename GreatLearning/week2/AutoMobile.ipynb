```python
label_encoder = LabelEncoder()
data['House_loan_code'] = label_encoder.fit_transform(data['House_loan'])
data['Partner_working_code'] = label_encoder.fit_transform(data['Partner_working'])
data['Make_code'] = label_encoder.fit_transform(data['Make'])
data['Gender_code'] = label_encoder.fit_transform(data['Gender'])
data['Profession_code'] = label_encoder.fit_transform(data['Profession'])
data['Marital_status_code'] = label_encoder.fit_transform(data['Marital_status'])
data['Education_code'] = label_encoder.fit_transform(data['Education'])

#Gender	Profession	Marital_status	
sns.heatmap(data=data[["Education_code","Marital_status_code","Profession_code","Gender_code","Personal_loan_code","Partner_working_code","House_loan_code"]].corr(), cmap='YlGnBu', annot=True, fmt='.2f')
```

```python
ct = pd.crosstab(data["Gender"],data["Make"])
ct.plot(kind="bar",stacked=True)
```

```python
ct = pd.crosstab(data["Profession"],data["Make"])
ct.plot(kind="bar",stacked=True)
```

```python
plt.figure(figsize=(12, 8)) 
ct = pd.crosstab(data["Price"],data["Age"])
ct.plot(kind="bar",stacked=True,figsize=(15, 5))
plt.show()
```

```python
plt.figure(figsize=(12, 8)) 
ct = pd.crosstab(data["Make"],data["Age"])
ct.plot(kind="bar",stacked=True,figsize=(15, 10))
plt.show()
```
