```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
tips =sns.load_dataset("tips")
```

```python
tips
```

```python
tips.describe()
```

```python
sns.displot(x=tips["size"],bins=4)
```

```python
sns.displot(x=tips["total_bill"],bins=20,kde=True)
```

```python
sns.boxplot(x=tips["total_bill"])
```

```python
sns.countplot(x=tips["sex"])
```

```python
plt.pie(tips["day"].value_counts(),labels=tips["day"].value_counts().index)
```

```python
sns.scatterplot(x=tips["total_bill"],y=tips["tip"])
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip")
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip")
```

```python
sns.boxplot(x=tips["day"],y=tips["total_bill"])
```
